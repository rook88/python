```python
# Simple Folder class that's attibutes are files. Setting an attribute writes file and getting an attribute reads file

import os

class Folder():
    def __init__(self, folder, encoding = 'Latin', expire_time = 3600):
        self._folder = folder
        self._encoding = encoding
        self._expire_time = expire_time
    def _file_path(self, name):
        return os.path.join(self._folder, name)
    def is_expired(self, name):
        return time.time() > os.path.getmtime(self._file_path(name)) + self._expire_time
    def __getattr__(self, name):
        try:
            with open(self._file_path(name), 'r', encoding = self._encoding, errors = 'replace') as f:
#                 try:
                ret = f.read()
#                 except UnicodeDecodeError:
#                     encoding = detectFileEncoding(os.path.join(self._folder, name))['encoding']
#                     with open(os.path.join(self._folder, name), 'r', encoding = encoding, errors = 'replace') as f:
#                         ret = f.read()
                return ret
        except FileNotFoundError:
            print("Not found")
            return None
    def __hasattr__(self, name):
        return os.path.isfile(self._file_path(name)) and not self.is_expired(name)
    def __setattr__(self, name, value):
        if name.startswith('_'):
            self.__dict__[name] = value
        else:
            with open(self._file_path(name), 'w', encoding = self._encoding) as f:
                f.write(value)
    def __getitem__(self, name):
        return self.__getattr__(name)
    def __setitem__(self, name, txt):
        self.__setattr__(name, txt)
    def __iter__(self):
         yield from [t for t in os.listdir(self._folder) if not self.is_expired(t)]
    def find(self, txt):
        return [name for name in self if txt.lower() in self.__getitem__(name).lower()]
    def items(self, grep = ''):
        return [x for x in self if grep in x]
    def edit(self, name = None):
        if name is not None:
            header = f"%%writefile {self._folder}/{name}"
            txt = self[name]
            if txt is None:
                txt = ''
        else:
            header = f"%%writefile {self._folder}/{_}"
            txt = ''
        get_ipython().set_next_input(header + "\n" + txt)
        

```

```python


class Duck():
    def quak(self):
        print("quak!")
    def __init__(self, name):
        self.name = name
        
class Dog():
    def bark(self):
        print("bark!")
    def __init__(self, name):
        self.name = name
        
def genCompositeObject(objects):
    class Ret():
        def __getattr__(self, name):
            for o in objects:
                if hasattr(o, name):
                    return getattr(o, name)
            raise AttributeError(f"composite object has no attribute '{name}'")
    return Ret()

duck = Duck('Donald')
dog = Dog('Goofy')
pet = genCompositeObject([duck, dog])

print(pet.name)
pet.bark()
pet.quak()
pet.pet()
```

```python
"""
Simple File class that's attibutes are parts of the. 
# Setting an attribute writes 
# getting an attribute reads fileparts
# if filepart is not found, None
"""

def fileCollectionAsDictionary(txt):
    ret = {}
    for item_raw in txt.split('#___'):
        item_raw_list = item_raw.split('___#\n')
        item_header = item_raw_list[0]
        try:
            item_txt = item_raw_list[1]
        except IndexError:
            continue
        ret[item_header] = item_txt
    return ret

def DictionaryAsFileCollection(d):
    ret = ''
    for name, item_txt in d.items():
        ret += '#___' + name + '___#\n' + item_txt
    return ret


import os
class FileCollection():
    def __init__(self, file):
        self.__dict__['_file'] = file
    def getFileDictionary(self):
        with open(self.__dict__['_file'], 'r') as f:
            fileTxt = f.read()
            return fileCollectionAsDictionary(fileTxt)
    def __getattr__(self, name):
        fileDict = self.getFileDictionary()
        if name in fileDict:
            return fileDict[name]
    def __setattr__(self, name, txt):
        fileDict = self.getFileDictionary()
        fileDict[name] = txt
        with open(self.__dict__['_file'], 'w') as f:
            f.write(DictionaryAsFileCollection(fileDict))
    def __getitem__(self, name):
        return self.__getattr__(name)
    def __setitem__(self, name, txt):
        self.__setattr__(name, txt)


demo = FileCollection('c:/temp/filepath')
# demo.bar
demo.bar2 = "new code"
```

```python
%%writefile tools.py
import sys
import os
import inspect
import importlib

__version__ = 1.01

class myFunction():
    def __init__(self, fun, tools):
        self.fun = fun
        self.tools = tools
    def __call__(self, *args, **kwargs):
        return self.fun(*args, **kwargs)
    def edit(self):
        self.tools.edit(self.fun.__name__)
    def edit(self):
        self.tools.clear(self.fun.__name__)

class Tools():
    def __init__(self, folder):
        self.__dict__['_folder'] = folder
        sys.path.append(folder)
    def filePath(self, name):
        return os.path.join(self.__dict__['_folder'], name) + '.py'
    def getCode(self, name):
        with open(self.filePath(name), 'r') as f:
            code = f.read()
        return code
    def edit(self, name):
        if name in sys.modules:
            with open(sys.modules[name].__file__, 'r') as f:
                code = f.read()
        elif os.path.isfile(self.filePath(name)):
            getCode(name)
        else:
            code = f"def {name}():\n    return '{name} called'"
        get_ipython().set_next_input(f"%%writefile {self.filePath(name)}\n{code}")
    def __getattr__(self, name):
        try:
            print(f"import {name}")
            __import__(name)
            importlib.reload(sys.modules[name])
            try:
                ret = myFunction(getattr(sys.modules[name], name), self)
            except AttributeError:
                print(f'Module {name} has no function {name}')
                self.edit(name)
                return None
            self.__dict__[name] = ret
            return ret
        except ModuleNotFoundError:
            print(f"{name} not found")
            self.edit(name)
            if name in self.__dict__:
                del self.__dict__[name]
    def commit(self, function):
        with open(self.filePath(function.__name__), 'w') as f:
            f.write(inspect.getsource(function))
        if function.__name__ in self.__dict__:
            del self.__dict__[function.__name__]
    def clear(self, name):
        del self.__dict__[name]
    def __iter__(self):
         yield from [t.replace('.py', '') for t in os.listdir(self.__dict__['_folder']) if not '__' in t]
    def find(self, txt):
        return [name for name in self if txt.lower() in self.getCode(name).lower()]
            
```

```python
import os
import getpass
import keyring
from cryptography.fernet import Fernet

def myEncrypt(txt):
    f = Fernet(keyring.get_password('secrets', 'folder').encode())
    return f.encrypt(txt.encode()).decode()


def myDecrypt(txt):
    f = Fernet(keyring.get_password('secrets', 'folder').encode())
    return f.decrypt(txt.encode()).decode()


class secretsFolder():
    def __init__(self, folder):
        self.__dict__['_folder'] = folder
    def __getattr__(self, name):
        filePath = os.path.join(self.__dict__['_folder'], name)
        if os.path.isfile(filePath):
            with open(filePath, 'r') as f:
                ret = f.read()
                return myDecrypt(ret)
        else:
            ret = getpass.getpass(f"File {filePath} does not exist. Give {name}")
            self.__setitem__(name, ret)
            return ret
    def __setattr__(self, name, txt):
        with open(os.path.join(self.__dict__['_folder'], name), 'w') as f:
            f.write(myEncrypt(txt))
    def __getitem__(self, name):
        return self.__getattr__(name)
    def __setitem__(self, name, txt):
        self.__setattr__(name, txt)


secrets = secretsFolder('g:/secrets')
```

```python
class Schema():
    def __init__(self, schema):
        self.__dict__['_schema'] = schema
    def __getattr__(self, table):
        return Table(self.__dict__['_schema'], table)
    def __getitem__(self, name):
        return self.__getattr__(name)

class Table():
    def __init__(self, schema, table):
        self.schema = schema
        self.table = table
    def __getitem__(self, columns):
        return Columns(self.schema, self.table, columns)
    def __call__(self, **where_kwargs):
        return genSelect(self.schema, self.table, ['*'], **where_kwargs)

class Columns():
    def __init__(self, schema, table, columns):
        self.schema = schema
        self.table = table
        self.columns = columns
    def __call__(self, **where_kwargs):
        return genSelect(self.schema, self.table, self.columns, **where_kwargs)
    
def genSelect(schema, table, columns, **where_kwargs):
    ret_select = f'select'
    ret_columns = '\n    ' + '\n   ,'.join(columns)
    ret_from = f'\nfrom\n    {schema}.{table}'
    ret_where = '\nwhere\n    1 = 1'
    for (key, value) in where_kwargs.items():
        ret_where += f'\nand {key} = {value}'   
    return ret_select + ret_columns + ret_from + ret_where + '\n;'
    
        
schemaFoo = Schema('schemaFoo')


assert schemaFoo.Tablebar() == 'select\n    *\nfrom\n    schemaFoo.Tablebar\nwhere\n    1 = 1\n;'
assert schemaFoo.Tablebar(col_2 = 4) == 'select\n    *\nfrom\n    schemaFoo.Tablebar\nwhere\n    1 = 1\nand col_2 = 4\n;'
assert schemaFoo.Tablebar['col_1', 'col_2'](col_1 = 1) == 'select\n    col_1\n   ,col_2\nfrom\n    schemaFoo.Tablebar\nwhere\n    1 = 1\nand col_1 = 1\n;'
assert schemaFoo['tableBar']() == 'select\n    *\nfrom\n    schemaFoo.tableBar\nwhere\n    1 = 1\n;'
# schemaFoo.Tablebar()
```

```python
def myVars(myObj):
    if isinstance(myObj, list):
        retList = []
        for subValue in myObj:
            retList.append(myVars(subValue)) 
        return retList
    if isinstance(myObj, Namespace):
        ret = {}
        for (key, value) in myObj.__dict__.items():
            ret.update({key : myVars(value)})
        return ret
    if hasattr(myObj, '__dict__'):
        ret = {}
        for (key, value) in vars(myObj).items():
            ret.update({key : myVars(value)})
        #return ret
        return {myObj.__class__.__name__ : ret}
    return myObj

from argparse import Namespace
def myNamespace(mydict):
    if not isinstance(mydict, dict):
        return mydict
    ret = Namespace(**{})
    for key, value in mydict.items():
        if isinstance(value, dict):
            setattr(ret, key, myNamespace(value))
        elif isinstance(value, list):
            setattr(ret, key, [myNamespace(subValue) for subValue in value])    
        else:
            setattr(ret, key, value)
    return ret

```

```python
import pickle

class PickleFolder():
    def __init__(self, folder, function = None):
        self.__dict__['_folder'] = folder
    def __hasattr__(self, name):
        return (str(name) in self.items())
    def __contains__(self, name):
        return self.__hasattr__(name)
    def __getattr__(self, name):
        with open(os.path.join(self.__dict__['_folder'], str(name)), 'rb') as f:
            return pickle.load(f)
    def __setattr__(self, name, obj):
        with open(os.path.join(self.__dict__['_folder'], str(name)), 'wb') as f:
            pickle.dump(obj, f)
    def __getitem__(self, name):
        return self.__getattr__(name)
    def __setitem__(self, name, obj):
        self.__setattr__(name, obj)
    def __iter__(self):
         yield from [t for t in os.listdir(self.__dict__['_folder'])]
    def items(self):
        return [x for x in self]
```

```python
import os

def initMyFolder(myFolder, path):
    myFolder.__dict__['_folderpath'] = path

def readFile(folderpath, name):
    with open(os.path.join(folderpath, name), 'r') as f:
        return f.read()

def writeFile(folderpath, name, txt):
    with open(os.path.join(folderpath, name), 'w') as f:
        f.write(txt)


myFolder = type('myFolder', (), {
    '__init__' : (lambda s, folder : initMyFolder(s, folder))
    ,'__getattr__' : (lambda s, name : readFile(s.__dict__['_folderpath'], name))
    ,'__setattr__' : (lambda s, name, txt : writeFile(s.__dict__['_folderpath'], name, txt))
})

w = myFolder('c:/temp')
w.bar = "some row before other rows\n" + w.foo + "and row after all rows"
```

```python
class myDataFrame():
    def __init__(self, df, name):
        self.df = df
        self.name = name
    def __add__(self, other):
        return myDataFrame(self.df.append(other.df), self.name + '_' + other.name)
    def __mul__(self, other):
        return myDataFrame(self.df.join(other.df, rsuffix = '_2'), self.name + '_j_' + other.name)
    def _repr_html_(self):
        return self.df.to_html()
    def __str__(self):
        return f"{self.name}, {self.df.shape[0]}x{self.df.shape[1]}"
    
        
def initMyFolder(myFolder, path):
    myFolder.__dict__['_folderpath'] = path

def readFile(folderpath, name):
    ret = pd.read_excel(os.path.join(folderpath, name + '.xlsx'))
    return myDataFrame(ret, name)

def writeFile(folderpath, name, df):
    df.to_excel(os.path.join(folderpath, name + '.xlsx'), index = False)

def deleteFile(folderpath, name):
    os.remove(os.path.join(folderpath, name + '.xlsx'))
    
def excelFiles(s):
    return [f.replace('.xlsx', '') for f in os.listdir(s.__dict__['_folderpath']) if f.endswith('.xlsx')]

def myIter(s):
    return iter([s.__getattr__(f) for f in excelFiles(s)])

def myContains(s, item):
    return item.name in excelFiles(s) 
    

myFolder = type('myFolder', (), {
    '__init__' : (lambda s, folder : initMyFolder(s, folder))
    ,'__getattr__' : (lambda s, name : readFile(s.__dict__['_folderpath'], name))
    ,'__setattr__' : (lambda s, name, df : writeFile(s.__dict__['_folderpath'], name, df))
    ,'__delattr__' : (lambda s, name : deleteFile(s.__dict__['_folderpath'], name))
    ,'__getitem__' : lambda s, name : s.__getattr__(name)
    ,'__setitem__' : lambda s, name, df : s.__setattr__(name, df)
    ,'__delitem__' : lambda s, name : s.__delattr__(name)
    ,'__contains__' : lambda s, item : myContains(s, item)
    ,'__iter__' : myIter
    ,'__len__' : lambda s : len(excelFiles(s))
})

excel = myFolder('c:/temp/excel')

print(len(excel))
# display(excel['df4'])
# del excel['df4']
len(excel)
print(excel.df3)
excel.df3 in excel
```

```python
# this is our descriptor object
class Bar(object):
    def __init__(self):
        self.value = ''
    def __get__(self, instance, owner):
        print(f"{self}, {instance}, {owner}")
        print("returned from descriptor object")
        return self.value * instance.value
    def __set__(self, instance, value):
        print("set in descriptor object")
        self.value = value
    def __delete__(self, instance):
        print("deleted in descriptor object")
        del self.value

class Foo(object):
    def __init__(self, value):
        self.value = value
    bar = Bar()

f = Foo(5)
f.bar = 10
print(f.bar)
# del f.bar
```
